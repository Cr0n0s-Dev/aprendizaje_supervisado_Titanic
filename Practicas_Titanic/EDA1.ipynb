```python
plt.figure(figsize=(12, 8))

sns.barplot(x = df['Survived'].value_counts().index, y = df['Survived'].value_counts())
plt.title('Value Counts of Survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
tasa_supervivencia = df['Survived'].value_counts()[1] / df['Survived'].value_counts().sum()
print(f'Tasa de supervivencia: {tasa_supervivencia:.2%}')
```

```python
# El promedio (.mean()) de una columna de 1s y 0s da exactamente la tasa de supervivencia
tasa_supervivencia = df['Survived'].mean()
print (tasa_supervivencia)
```

```python
# El secreto está en 'hue': divide las barras de la clase según si sobrevivieron o no
sns.countplot(x="Pclass", hue="Survived", data=df)

# Detalles finales para que se vea ordenado
plt.title("Tasa de Supervivencia por Clase")
plt.xlabel("Clase del Pasajero")
plt.ylabel("Cantidad de Pasajeros")
plt.show()
```

```python
# El secreto está en 'hue': divide las barras de la clase según si sobrevivieron o no
sns.countplot(x="Sex", hue="Survived", data=df)

# Detalles finales para que se vea ordenado
plt.title("Tasa de Supervivencia por Sexo")
plt.xlabel("Sexo del Pasajero")
plt.ylabel("Cantidad de Pasajeros")
plt.show()
```

```python
plt.figure(figsize = (12, 8))

# colores para cada una de las clases
color = ['#240750', '#344C64']

sns.countplot(df, x = 'Embarked', hue = 'Survived', palette=color) 
plt.legend(title = 'Supervivencia', labels = ['No sobrevivió', 'Sí sobrevivió'])
plt.show()
```

```python
plt.figure(figsize = (12, 8))

sns.histplot(df, x = 'Age', kde = True)
plt.title('Distribución de Edades')
plt.axvline(df['Age'].mean(), color = 'red')
plt.axvline(df['Age'].median(), color = 'green')
plt.axvline(df['Age'].mode()[0], color = 'black')
plt.show()

# Print the results
print(f"Media: {df['Age'].mean()}")
print(f"Mediana: {df['Age'].median()}")
print(f"Moda: {df['Age'].mode()[0]}")
```
