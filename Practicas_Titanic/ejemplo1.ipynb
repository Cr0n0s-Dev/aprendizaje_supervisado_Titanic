```python
import pandas as pd
df = pd.read_csv('./data/titanic.csv')
df.sample(5)
```

```python
print(f"El shape del conjunto de entrenamiento es {df.shape}. {df.shape[0]} filas y {df.shape[1]} columnas.") 
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Crear el mapa de calor de valores nulos
# df.isnull() genera una tabla de True (si es nulo) y False (si tiene datos)
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap="viridis")

# 2. Añadir título y ajustar etiquetas para que sea legible
plt.title("Mapa de Calor de Valores Nulos", fontsize=16)
plt.xlabel("Columnas / Características", fontsize=12)
plt.ylabel("Filas / Registros", fontsize=12)

# 3. Guardar el gráfico como imagen (buena práctica en scripts de producción)
plt.savefig("mapa_de_calor_nulos.png", bbox_inches="tight")
```

```python
print(df.isnull().sum())
```

```python
# Código directo
porcentaje_nulos = (df.isnull().sum() / len(df) * 100).round(2).sort_values(ascending=False)

# Mostrar el resultado
print(porcentaje_nulos)
# Esto solo cambia la forma en que se imprime en pantalla
print(porcentaje_nulos.map("{:.2f}%".format))
```
